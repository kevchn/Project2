```python
""" Linear SVM on MNIST with PCA """

#%%
from sklearn.datasets import fetch_mldata
from sklearn import svm
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from timeit import default_timer as timer
```

```python
# Dataset
mnist = fetch_mldata('MNIST original', data_home="../data/raw")
X = mnist.data
y = mnist.target
```

```python
# Scale to [-1, 1]
X = X/255.0*2 - 1
```

```python
# Dimensionality reduction
pca = PCA(n_components=16, svd_solver='randomized', whiten=True)
X_transformed = pca.fit_transform(X)

plt.hist(pca.explained_variance_ratio_, bins=16, log=True)
pca.explained_variance_ratio_.sum()
```

```python
fig, plot = plt.subplots()
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y)
plt.tight_layout()
plt.title('MNIST after PCA (k=2)')
plt.show()
```
